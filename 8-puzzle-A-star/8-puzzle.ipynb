```python
class position:
    def __init__(self,puz,f_cost,g_cost):
        self.puz = puz
        self.f_cost = f_cost
        self.g_cost = g_cost
        self.parent = None
        self.direction = None
    def find_element(self,element):
        for i in range(len(self.puz)):
            for j in range(len(self.puz)):
                if self.puz[i][j] == element:
                    return i,j
                
    def check_possible_moves(self,lis):
        possible_moves = []
        for i in lis:
            if i[0] >= 0 and i[0] < len(self.puz) and i[1] >= 0 and i[1] < len(self.puz):
                possible_moves.append(i)
        return possible_moves
    
    def child_nodes(self):
        x, y = self.find_element('_')
        possible_directions = self.check_possible_moves([[x-1,y],[x+1,y],[x,y-1],[x,y+1]])
        children = []
        for i in possible_directions:
            child = self.move(x,i[0],y,i[1])
            if child is not None:
                child_position = position(child,0,(self.g_cost + 1))
                children.append(child_position)
        return children
    
    def make_copy(self,origin):
        copy_puz = []
        for i in origin:
            temp = []
            for j in i:
                temp.append(j)
            copy_puz.append(temp)
        return copy_puz
    
    def move(self,x1,x2,y1,y2):
        child_puz = []
        new_puz = self.make_copy(self.puz)
        temp = new_puz[x1][y1]
        new_puz[x1][y1] = new_puz[x2][y2]
        new_puz[x2][y2] = temp
        return new_puz
    
class puzzle:
    def __init__(self,size):
        self.size = size
        self.open_list = []
        self.close_list = []
        
    def get_puzzle(self):
        puzzle = []
        for i in range(self.size):
            puzzle.append(input().split(" "))
            
        for i in range(self.size):
            for j in range(self.size):
                if puzzle[i][j] != '_':
                    puzzle[i][j] = int(puzzle[i][j])
                    
        return puzzle
    
    def h(self,first,second):
        h_cost = 0
        for i in range(self.size):
            for j in range(self.size):
                if first[i][j] != '_' and first[i][j] != second[i][j]:
                    h_cost = h_cost + 1
        return h_cost
    
    def f(self,first,second):
        return self.h(first.puz,second)+first.g_cost
    
    def print_puz(self,puzz):
        puz = puzz.puz
        print("***************")
        for i in puz:
            for j in i:
                print(j,end=" ")
            print("")
        if puzz.direction is not None:
            print("direction : ",puzz.direction)
        print("f_cost is : ",puzz.f_cost)
        
    def direction(self,new,old):
        new_x,new_y = new.find_element('_')
        old_x,old_y = old.find_element('_')
        if new_x == old_x -1:
            new.direction = "up"
        if new_x == old_x + 1:
            new.direction = "down"
        if new_y == old_y - 1:
            new.direction = "left"
        if new_y == old_y + 1:
            new.direction = "right"
    def find_path(self,cur):
        path = []
        s = cur
        while s.parent is not None:
            self.direction(s,s.parent)
            path.append(s)
            s = s.parent
        path.append(s)
        path.reverse()
        for i in path:
            self.print_puz(i)
            
    def solver(self):
        print("please enter start puzzle: \n")
        start = self.get_puzzle()
        print("please enter goal puzzle:\n")
        goal = self.get_puzzle()
        start = position(start,0,0)
        start.f_cost = self.f(start,goal)
        self.open_list.append(start)
        while len(self.open_list) > 0 :
            cur = self.open_list[0]
            if self.h(cur.puz,goal) == 0:
                self.find_path(cur)
                break
            children = cur.child_nodes()
            for i in children:
                flag=1
                i.f_cost = self.f(i,goal)
                i.parent = cur
                for j in range(len(self.open_list)):
                    if self.open_list[j].puz == i.puz and self.open_list[j].f_cost <= i.f_cost:
                        flag=0
                for j in range(len(self.close_list)):
                    if self.close_list[j].puz == i.puz and self.close_list[j].f_cost <= i.f_cost:
                        flag=0
                if flag:
                    self.open_list.append(i)
                    
            self.close_list.append(cur)
            b = self.open_list.pop(0)
            self.open_list.sort(key=lambda x:x.f_cost , reverse=False)
            
n = int(input("For a n*n puzzle,please enter n : "))            
puzzl = puzzle(n)
puzzl.solver()
```
